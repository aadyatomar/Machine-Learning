```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = list(stopwords.words('english'))
stop_words.extend(["It","know","In","For","day","must", "And", "But", "like","came","still","find","let","Let","To","ask","shall","always","go","make","In"])

word_tokens1 = word_tokenize(S1)
#they are present in stop_words or not
filtered_sentence1 = [w for w in word_tokens1 if not w.lower() in stop_words]
#with no lower case conversion
filtered_sentence1 = []
 
for w in word_tokens1:
    if w not in stop_words:
        filtered_sentence1.append(w)
 
print(word_tokens1)

```

```python
print(filtered_sentence1)
```
